**Publishing & Sharing Reports in Power BI**

1  . Publishing and sharing reports in power bi desktop is considered more
     important than just building deshboards because reports enable intractive detailed data analysis and shorytelling that can ba consumed
     by stakeholders sharing them makes the insights accessible actionable and collaborative across the organization driving data driven decisions .

2  . Power bi service is the cloude based online platfrom where reports and

3  . Publishing a reports in power bi means uploading the desktop file to
     the power bi service so it can be viewed and intracted with online .two outcomes after publishing are . the report becomes accessible to authorized users via web or mobile .
     the dataset can be set up for scheduled refresh to keep data current .

4  . Steps to publish a report from power bi desktop to power bi service
   . save the .pbix file .
   . click public in the home ribbon
   . choose the traget workspace in power bi service .
   . authenticate with your power bi account if prompted

5  . In power bi a report is a multi-page collection of visualizations built
     on a dataset focused on detailed analysis and storytelling . A dashboard is a single-page canvas of pinned visual from one or more reports designed for quick monitoring and high-level insights .
   
6  . Direct report sharing - share a specific report with individuals or
     groups permissions can be set for view
   . Workspace access - grant users access to an entire workspace permissions use case give a team full access to a project analytics workspace .
   . Power bi apps - publish curated contain as an app for broader distribution with controlled consumption experience.

7  . Admin  - Full control over the workspace
   . Member - Can view edit and publish contain in the workspace and add
              other members .
   . Contributor - Can create edit and publish content but cannot manage
                   workspace settings or add members .
   . viewer - Can only view published reports and dashboards on editing
              rights .

8  . Governance  - Apps provide a centralized version controlled
                   distribution channel for curated reports enforcing consistant access policies and audit trails .
   . Enterprise usage - Apps decouple consmption from workspace permission
                       allowing administrators to manage large-scale sharing enforce row-level security and apply service level agreements without exposing raw workspace access to end users .

9  . RLS ensure different users see different data in the same report by
     applying DAX filters to the data model thet evaluate the users identity and hide unauthorized rows delivering personalized views without needing separate reports .

10  . Data leakage - unrestricted access can expose sensitive data to
      unauthorized users . use RSL and apps to enforce data level permissions.
   . Governance gaps - unmanaged sharing leads to inconsistent versioning
     and audit issues .
   . Misuse of insights - lack of role based access can cause improper data
     interpretation or unauthorized distribution .
